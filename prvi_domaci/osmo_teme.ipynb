```python
import numpy as np
```

```python
def Afinize(tacka):
    x1 = tacka[0]/tacka[2]
    x2 = tacka[1]/tacka[2]
    x3 = 1
    return [x1,x2,x3]
```

```python
def prosek(t,p,q):
    x1 = (t[0] + p[0] + q[0]) / 3
    x2 = (t[1] + p[1] + q[1]) / 3
    x3 = (t[2] + p[2] + q[2]) / 3
    return [x1,x2,x3]
```

```python
def zavrsno(A):
    X = []
    X.append(np.round(A[0]))
    X.append(np.round(A[1]))
    return X
```

```python
def hom2aff(points):
    X = []
    for point in points:
        X.append([point[0], point[1]])
    return X
```

```python
def osmoteme(tacke):
    Xb1 = Afinize(np.cross(np.cross(tacke[1], tacke[4]), np.cross(tacke[2], tacke[5])))
    Xb2 = Afinize(np.cross(np.cross(tacke[1], tacke[4]), np.cross(tacke[0], tacke[3])))
    Xb3 = Afinize(np.cross(np.cross(tacke[0], tacke[3]), np.cross(tacke[2], tacke[5])))

    Xb = prosek(Xb1,Xb2,Xb3)

    Yb1 = Afinize(np.cross(np.cross(tacke[0], tacke[1]), np.cross(tacke[3], tacke[4])))
    Yb2 = Afinize(np.cross(np.cross(tacke[0], tacke[1]), np.cross(tacke[5], tacke[6])))
    Yb3 = Afinize(np.cross(np.cross(tacke[3], tacke[4]), np.cross(tacke[5], tacke[6])))

    Yb = prosek(Yb1,Yb2,Yb3)

    Zb1 = Afinize(np.cross(np.cross(tacke[1], tacke[2]), np.cross(tacke[4], tacke[5])))
    Zb2 = Afinize(np.cross(np.cross(tacke[1], tacke[2]), np.cross(tacke[3], tacke[6])))
    Zb3 = Afinize(np.cross(np.cross(tacke[3], tacke[6]), np.cross(tacke[4], tacke[5])))

    Zb = prosek(Zb1,Zb2,Zb3)
    
    P41 = Afinize(np.cross(np.cross(Xb, tacke[6]), np.cross(Yb, tacke[2])))
    P42 = Afinize(np.cross(np.cross(Xb, tacke[6]), np.cross(Zb, tacke[0])))
    P43 = Afinize(np.cross(np.cross(Yb, tacke[2]), np.cross(Zb, tacke[0])))

    P4 = prosek(P41, P42, P43)
    P4 = zavrsno(P4)

    return P4
```

```python
tacka4 = osmoteme(tacke)
tacke = hom2aff(tacke)
print(tacke)
print(tacka4)
```
